Is a method to model the realtionship between one independent variable(X) and one dependent variable(Y) using a straight line.

```python
# Importing Libraries
#pandas-
#numpy-numerical operators performance
#plt- for visualization
#LinearRegression- We are doing simplelinear regression so we are importing that.
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Sample dataset: Temperature vs Ice Cream Sales
np.random.seed(0)
temperature = np.random.randint(20,40,50)
sales = 2.5*temperature + np.random.normal(0,5,50)
df = pd.DataFrame({'Temperature': temperature, 'IceCreamSales': sales})
print(df)
```

```python
# Slecting the model
X = df[['Temperature']]    # x-axis is temperature in straight line.
Y = df['IceCreamSales']    # y-axis is icecreamsales in straight line.
model = LinearRegression() # we are selecting model that is linearregression.
model.fit(X,Y)
```

```python
# Plotting
plt.scatter(X,Y,color='blue') # It will draw only dots and it's color is blue.
plt.plot(X,model.predict(X),color='red') # we are drawing a plot(only redline).
plt.xlabel("Temperature (C)") # on x-axis temperature.
plt.ylabel("Ice Cream Sales") # on y-axis icecreamsales.
plt.title("Simple Linear Regression") # title
plt.show() # showing.
```
