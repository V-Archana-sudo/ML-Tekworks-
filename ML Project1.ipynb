```python
# Mini Capstone: predict student scores
#This notebook demonstrates how to build a simple linear regression model to predict student scores based on hours studied.
```

```python
# Step-01: Importing Libraries
import warnings
warnings.filterwarnings("ignore")
import pandas as pd # Creating dataframes
import matplotlib.pyplot as plt # For visulization
from sklearn.model_selection import train_test_split # We split the data into two parts:train,test.
from sklearn.linear_model import LinearRegression # We are applying simplelinearregression so we implemented it.
from sklearn.metrics import mean_absolute_error,r2_score # We can able to see accuracy of the model.
```

```python
# Step-02: Load DataSet
df = pd.read_csv("https://raw.githubusercontent.com/AdiPersonalWorks/Random/master/student_scores%20-%20student_scores.csv")
df.head()
```

```python
# Step-03: Visualize Data
plt.scatter(df["Hours"],df["Scores"]) # It will only display dots with blue color.
plt.title("Hours vs Scores") # Title
plt.xlabel("Hours Studied") # x-axis
plt.ylabel("Score") # y-axis
plt.grid(True) # It will give that grid boxes in the graph.....
plt.show()
```

```python
# Step-04: Prepare Data
X = df[["Hours"]]
Y = df["Scores"]
# We will prepare data to split it.
```

```python
#Step-05: Splitting the data into test,train two parts (Train/Test split)
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
# We are splitting the x-axis data to train and test. similarly splitting the y-axis data to train and test.
# We will have train and test for both X and Y So (X_train,X_test,Y_train,Y_test)
# test_size=0.2 (Means 20% for test and 80% for training).
# random_state=42 (Means we will take 42 randomly in both X and Y).
```

```python
#Step-06: Applying model that is simplelinear regression nothing but linear regression(Train Linear Regression model)
model = LinearRegression() # We are using linearregression.
model.fit(X_train,Y_train) # We are fitting the model for X_train and Y_train.
# model.fit(X_train,Y_train) means we are fitting or training the train data which is in x-axis. similarly we are training the train data which is in y-axis.
```

```python
# Step-07: Predict
# We want Y_test value as output. so we are prediciting the Y-test value using X_test as input.
Y_pred=model.predict(X_test)
```

```python
# Step08: Evaluate model (On predict value(y_pred) and original ones(y_test)).
print("R^2 Score:" , r2_score(Y_test,Y_pred))
print("Mean Absolute Error:", mean_absolute_error(Y_test,Y_pred))
```

```python
#R^2 Score: 0.9678055545167994 means 96% Your model is accurate.
```

```python
# Step-09: Visualize Regression line
plt.scatter(X,Y,color='blue')
plt.plot(X,model.predict(X),color='red') # Regression line means that red line.
plt.title("Regression Line - Student Scores")
plt.xlabel("Hours Studied")
plt.ylabel("Score")
plt.show()
```

```python
# Step-10: predicting score based on the study hours or (predict for custom input)
hours=[[9.25]]
predicted_score = model.predict(hours)
print(f"predicted score for studying {hours[0][0]} hours = {predicted_score[0]:.2f}")
```
